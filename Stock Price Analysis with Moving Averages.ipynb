## ðŸ“Š Stock Price Analysis with Moving Averages

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Take stock symbol input (Example: RELIANCE.NS, AAPL)
ticker = input("Enter the stock symbol: ")

# Download stock data for 2023
df = yf.download(ticker, start="2023-01-01", end="2024-01-01", interval="1d")
```

```python
# Calculate 7-day Moving Average (short-term trend)
df["7_day_MA"] = df["Close"].rolling(window=7).mean()

# Calculate 30-day Moving Average (long-term trend)
df["30_day_MA"] = df["Close"].rolling(window=30).mean()
```

```python
# Plot Closing Price and Moving Averages
plt.figure(figsize=(10,5))

plt.plot(df["Close"], label="Close Price")      # Actual price
plt.plot(df["7_day_MA"], label="7-Day MA")     # Short-term trend
plt.plot(df["30_day_MA"], label="30-Day MA")   # Long-term trend

plt.title(f"{ticker} Stock Price Analysis")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)

plt.show()
```
